```python
#Print all numbers between two numbers entered by the user.

a=int(input("enter a number:"))
b=int(input("enter b number:"))
i=0
if a<b:
  i=a+i
  while i<b:
    print(i)
    i+=1
elif a>b:
  i=b+i
  while i<a:
    print(i)
    i+=1
else:
  print('both number is equal.')
```
